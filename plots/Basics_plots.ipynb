# Introduction

```python
import pandas as pd
import matplotlib as mtl

%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
print(recent_grads.shape[0])
recent_grads["Sample_size"].value_counts()
```

# Scatter Plots

```python
recent_grads.plot(kind="scatter",x="Sample_size",y="Median")
recent_grads.plot(kind="scatter",x="Sample_size",y="Unemployment_rate")
recent_grads.plot(kind="scatter",x="Full_time",y="Median")
recent_grads.plot(kind="scatter",x="ShareWomen",y="Median")
recent_grads.plot(kind="scatter",x="Men",y="Median")
recent_grads.plot(kind="scatter",x="Total",y="Median")
```
